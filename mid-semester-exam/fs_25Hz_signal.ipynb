```python
import numpy as np
import matplotlib.pyplot as plt

# Signal parameters
T = 1
Fs = 25

# Generate time array and signal
Ts = 1 / Fs
t = np.arange(0, T, Ts)
y = np.sin(2 * np.pi * 8 * t)

# Plot the signal
label = f'{Fs} Hz Sine Wave'
plt.plot(t, y, label=label)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title(label)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Signal parameters
T = 1
Fs_arr = [25, 50, 100, 200]

# Generate time array and signal

Ts = 1 / Fs  # Sampling period
t = np.arange(0, T, Ts)  # Time array

plt.figure(figsize=(10, 6))
for Fs in Fs_arr:
    y = np.sin(2 * np.pi * 8 * t)  # Sine wave signal at 8 Hz
    label = f'{Fs} Hz Sampling Frequency'
    plt.plot(t, y, label=label)  # Plot signal


# Plot the signal
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Sine Waves at Different Sampling Frequencies')
plt.grid(True)
plt.legend()
plt.show()
```
